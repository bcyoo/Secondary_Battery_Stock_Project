```python
# 2차전지 csv 파일 불러오기
df = pd.read_csv('./data/new_secondary_cell_pluscode_wonoise_re.csv', engine='python')
```

```python
# datetime을 이용해 출원날짜 데이터 형식 변경
df['출원일자'] = pd.to_datetime(df['출원일자'])

# '출원년월' column 생성
df.insert(3,'출원년월', 0, allow_duplicates=False)

# '출원년월' column에 %Y-%m 형식으로 변경
df['출원년월'] = df['출원일자'].dt.strftime('%Y-%m')
df
```

```python
# 데이터프레임 저장 (.gz를 통해 압축저장 가능)
df.to_csv(f'./data/new_secondary_cell_pluscode_wonoise_re.csv', index=False, encoding='utf-8-sig')
```
